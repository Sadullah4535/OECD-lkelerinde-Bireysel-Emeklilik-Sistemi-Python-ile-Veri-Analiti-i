# Kümeleme Analizi

K-Means kümeleme algoritması, gözlemleri önceden belirlenmiş sayıda küme veya gruba ayırmak için kullanılan yaygın bir kümeleme yöntemidir. İşte algoritmanın çalışma prensibi ve optimum küme sayısının nasıl belirleneceği hakkında daha fazla ayrıntı:

1. K-Means Kümeleme Prensibi
Adım 1: K rastgele küme merkezi (centroid) seçilir.
Adım 2: Her veri noktası, en yakın küme merkezine atanır.
Adım 3: Her küme için, yeni küme merkezleri mevcut kümeye ait veri noktalarının ortalaması alınarak güncellenir.
Adım 4: Küme merkezleri değişene kadar veya önceden belirlenmiş bir kriter sağlanana kadar 2. ve 3. adımlar tekrarlanır.
2. n_clusters Parametresi
n_clusters parametresi, oluşturmak istediğiniz küme sayısını belirler. Bu değerin seçimi, kümeleme analizinin başarısını doğrudan etkiler. Yanlış bir küme sayısı, kümelerin kalitesini düşürebilir.

3. Optimum Küme Sayısının Belirlenmesi: Dirsek Yöntemi (Elbow Method)
Optimum küme sayısını belirlemek için "dirsek yöntemi" (Elbow Method) kullanılır. Bu yöntem, her bir küme sayısı için WCSS'yi (Within-Cluster Sum of Squares) hesaplar ve küme sayısının artmasına rağmen WCSS değerindeki azalmanın belirgin bir şekilde yavaşladığı noktayı bulur.

PCA, genellikle 2 veya 3 bileşen elde edilene kadar kullanılır, bu da veri kümesinin daha kolay görselleştirilmesini sağlar.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Örnek veri (kendi veri setinizi df olarak yükleyin)
df

# Özellikleri seçin (sadece sayısal sütunları kullanmak genellikle en iyi yöntemdir)
X = df.select_dtypes(include='number')

# PCA kullanarak boyut indirgeme (2 bileşene indirgeme)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# K-Means kümeleme algoritmasını kullanarak kümeleri belirleme
kmeans = KMeans(n_clusters=2, random_state=42)  # 2 kümeyi örnek olarak seçtik
df['Cluster'] = kmeans.fit_predict(X)

# PCA ile indirgenmiş veriyi DataFrame'e dönüştürme
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
df_pca['Cluster'] = df['Cluster']

# Kümeleri görselleştirme
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_pca, palette='viridis', s=100)
plt.title("PCA ile Boyut İndirgeme ve K-Means Kümeleme")
plt.xlabel("PCA Bileşeni 1")
plt.ylabel("PCA Bileşeni 2")
plt.legend()
plt.show()
```

```python
# Kümelerin merkezlerini hesaplama
cluster_centers = kmeans.cluster_centers_
pca_centers = pca.transform(cluster_centers)  # Kümelerin merkezlerini PCA uzayına projekte et

# Kümeleri görselleştirme
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_pca, palette='viridis', s=100)
plt.scatter(pca_centers[:, 0], pca_centers[:, 1], c='red', s=200, marker='X', label='Merkezler')

# Kümelerin merkezlerine etiket ekleme
for i, (x, y) in enumerate(pca_centers):
    plt.text(x, y, f"Küme {i}", color='black', fontsize=10, ha='center', va='center')

plt.title("PCA ile Boyut İndirgeme ve K-Means Kümeleme (Etiketli)")
plt.xlabel("PCA Bileşeni 1")
plt.ylabel("PCA Bileşeni 2")
plt.legend()
plt.show()
```

Kümelerin merkezleri kırmızı 'X' işaretiyle gösterildi, ve her merkeze "Küme 0", "Küme 1", vb. şeklinde etiketler eklendi.

# PCA Sonrası Yorumlama
Eğer PCA ile indirgenmiş veride kümeler iyi ayrılmamışsa, veri setinin yapısına göre başka yöntemler (örneğin, t-SNE, UMAP) denenebilir.

PCA sonrası veriyi yorumlamak oldukça önemlidir. PCA, verideki en fazla varyansı içeren bileşenleri bulmaya çalışır, ancak bu her zaman kümelerin net bir şekilde ayrışmasını sağlamayabilir. Böyle bir durumda aşağıdaki yöntemleri kullanmak daha yararlı olabilir:

Alternatif Yöntemler
t-SNE (t-Distributed Stochastic Neighbor Embedding)

Özellikle yüksek boyutlu veri setlerinde kullanılabilir.
t-SNE, veriyi düşük boyutlara indirgerken, yakın komşuluk ilişkilerini koruyarak kümelerin görselleştirilmesini sağlar.
Bu yöntem, özellikle karmaşık ve çok boyutlu veri setleri için uygundur ve kümeler arasındaki ilişkileri daha net ortaya çıkarabilir.
UMAP (Uniform Manifold Approximation and Projection)

t-SNE'ye benzer, ancak genellikle daha hızlıdır ve büyük veri setleri üzerinde etkili çalışır.
UMAP, verinin topolojik yapısını koruyarak boyut indirgeme yapar ve kümelerin ayrışmasını kolaylaştırır.
Yorumlama ve Analiz
Kümeler İyi Ayrılmadıysa: PCA ile indirgenen veride kümeler net bir şekilde ayrışmıyorsa, bu, verinizin yapısının daha karmaşık olduğu anlamına gelebilir. Örneğin, veri setinizde kümelerin doğrusal olmayan ilişkilerle birbirinden ayrılması gerekiyorsa, PCA yeterince etkili olmayabilir.
Veri Seti Yapısı: PCA'nın başarısı, verinin varyansını iyi temsil eden bileşenlerin olup olmamasına bağlıdır. Eğer verinizde çok sayıda gürültü (anlamsız bilgi) veya doğrusal olmayan ilişkiler varsa, PCA bu ilişkileri yeterince yakalayamayabilir.

# t-SNE 

t-SNE (t-Distributed Stochastic Neighbor Embedding), yüksek boyutlu veriyi 2 veya 3 boyutlu bir uzaya indirgerken, benzer verileri birbirine yakın, farklı verileri ise uzak tutmaya çalışan bir tekniktir. Özellikle kümelerin görselleştirilmesinde çok etkili bir yöntemdir, çünkü t-SNE, verinin topolojik yapısını ve komşuluk ilişkilerini koruyarak düşük boyutlu bir projeksiyon sağlar.

Aşağıda, t-SNE kullanarak verinizi 2 boyutlu bir uzaya indirgeme ve kümeleri görselleştirme örneği yer alıyor:

Adımlar
Verinizi seçin ve boyut indirgeme yapın.
Kümeleri belirlemek için t-SNE algoritmasını uygulayın.
Sonuçları görselleştirin

```python
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns


# Özellikleri seçme (sayısal sütunları kullanmak en iyi sonucu verir)
X = df.select_dtypes(include='number')

# t-SNE ile boyut indirgeme (2 bileşenli)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# KMeans kümeleme (önceden belirlediğiniz küme sayısını kullanın)
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# t-SNE ile indirgenmiş veriyi DataFrame'e dönüştürme
df_tsne = pd.DataFrame(X_tsne, columns=['tSNE1', 'tSNE2'])
df_tsne['Cluster'] = df['Cluster']

# Kümeleri görselleştirme
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tSNE1', y='tSNE2', hue='Cluster', data=df_tsne, palette='magma', s=100)
plt.title("t-SNE ile Boyut İndirgeme ve K-Means Kümeleme")
plt.xlabel("t-SNE Bileşeni 1")
plt.ylabel("t-SNE Bileşeni 2")
plt.legend()
plt.show()
```

t-SNE'nin Özellikleri:
Komşuluk İlişkilerini Korur: t-SNE, benzer verilerin birbirine yakın, farklı verilerin ise uzak olmasını sağlar.
Büyük Veri Setlerinde Zorlayıcı: t-SNE büyük veri setlerinde zaman alıcı olabilir, bu nedenle küçük veri setlerinde daha iyi sonuçlar verebilir.
Çok Boyutlu Veriyi Anlamlandırma: Özellikle doğrusal olmayan ilişkileri görselleştirmek için etkilidir.

Eğer verinin yapısı daha karmaşıksa (çok boyutlu ve doğrusal olmayan ilişkiler içeriyorsa), t-SNE çok faydalı olabilir.
