# Uygun Küme Sayısı
Uygun küme sayısını belirlemek için dirsek yöntemi (elbow method) ve silüet yöntemi (silhouette method) gibi teknikleri kullanabiliriz. Bu yöntemler, verilerin küme sayısına göre nasıl gruplandığını görmemize ve doğru küme sayısını belirlememize yardımcı olur.

Aşağıda her iki yöntemi de uygulayan bir kod örneği bulunmaktadır:

# 1. Dirsek Yöntemi (Elbow Method)
Dirsek Yöntemi, bir kümeleme algoritmasının hata kareleri toplamını (inertia) farklı küme sayıları için hesaplar ve bu değerlerin grafik üzerinde nasıl değiştiğini inceler. Küme sayısını seçerken, inertia'nın azalma hızının yavaşladığı "dirsek" noktasına bakılır. Bu noktada, küme sayısı artarken eklenen kümelerin, toplam hata kareleri toplamını çok fazla azaltmadığı görülür.
Kümelerin merkezleri (centroids).
Yöntem, uygun küme sayısını belirlerken, inertia değerinin azalma hızının dramatik şekilde yavaşladığı noktayı seçer.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np


# DataFrame oluştur
df = pd.DataFrame(data)

# Yıllara göre veriyi seç ve 'X' değişkenine atayalım
X = df[["Pandemi Öncesi Ortalama", "Pandemi Sonrası Ortalama"]]

# Küme sayısı aralığı (1 ile 10 arasında küme sayıları)
K_range = range(1, 10)


# **Dirsek Yöntemi (Elbow Method)**
ssd = []  # Her bir küme sayısına karşılık gelen toplam hata kareleri toplamı (inertia)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    ssd.append(kmeans.inertia_)

# Dirsek grafiğini çizme
plt.figure(figsize=(10, 5))
plt.plot(K_range, ssd, marker='o', linestyle='--')
plt.xlabel('Küme Sayısı (k)')
plt.ylabel('Toplam Hata Kareleri (Inertia)')
plt.title('Dirsek Yöntemi ile Uygun Küme Sayısını Belirleme')
plt.show()

# Dirsek yöntemine göre uygun küme sayısını yazdırma
# "dirsek" noktası, değişimin hızla yavaşladığı küme sayısına işaret eder.
# Bu noktayı görsel olarak veya matematiksel olarak bulabiliriz. 
# Burada görsel inceleme yapalım ve k değeri yazdıralım.


optimal_k_elbow = K_range[np.argmax(ssd) + 1]  # Küme sayısı 1'den başlar, diff 0'dan başlar
print(f"Dirsek Yöntemi ile optimal küme sayısı: {optimal_k_elbow}")
```
