# Kümeleme Analizi
Ülkeleri BES varlıklarına göre gruplamak için kümeleme analizi.

Öncelikle, K-Means kümeleme analizinde ideal küme sayısını belirlemek için Elbow (Dirsek) yöntemini kullanabiliriz. Bu yöntemde, farklı küme sayılarında (örneğin 1'den 10'a kadar) model oluşturulur ve her birinin "inertia" değeri hesaplanır. Bu değerlerin grafiği çizilerek, bir noktadan sonra artışın azaldığı (dirsek oluşturduğu) yerdeki küme sayısı seçilir. Bu adımları tamamladıktan sonra, belirlediğimiz küme sayısıyla K-Means algoritmasını uygulayacağız ve ülkeleri etiketleyeceğiz.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Veri setini yükle
data = {
    "Ülkeler": ["ABD", "Kanada", "Birleşik Krallık", "Avustralya", "Japonya", 
                "Hollanda", "İsviçre", "Danimarka", "İsveç", "Kore", 
                "İsrail", "Fransa", "Almanya", "Meksika", "İtalya", 
                "Şili", "Belçika", "Finlandiya", "İspanya", "İrlanda", 
                "Yeni Zelanda", "İzlanda", "Polonya", "Portekiz", "Çekya", 
                "Türkiye", "Slovakya Cum.", "Macaristan", "Letonya", "Litvanya", 
                "Estonya", "Slovenya", "Yunanistan", "Lüksemburg"],
    "log2018": [7.44, 6.40, 6.42, 6.27, 6.23, 6.19, 6.00, 5.83, 5.67, 5.66, 
                 5.31, 5.42, 5.43, 5.29, 5.30, 5.29, 5.26, 5.18, 5.24, 5.10, 
                 4.74, 4.59, 4.68, 4.65, 4.32, 0.00, 4.08, 3.90, 3.67, 3.57, 
                 3.70, 3.56, 3.20, 3.27],
    "log2019": [7.51, 6.45, 6.50, 6.26, 6.24, 6.25, 6.06, 5.88, 5.73, 0.00, 
                 5.41, 5.44, 5.46, 5.38, 5.34, 5.33, 5.27, 5.21, 5.26, 5.17, 
                 4.82, 4.64, 4.68, 4.69, 4.35, 0.00, 4.12, 3.93, 3.76, 3.65, 
                 3.77, 3.60, 3.24, 3.31],
    "log2020": [7.56, 6.49, 6.55, 6.25, 6.26, 6.32, 0.00, 5.95, 5.82, 5.75, 
                 5.48, 5.50, 5.53, 5.42, 5.41, 5.32, 0.00, 5.27, 5.30, 5.21, 
                 4.90, 4.68, 4.69, 4.74, 4.40, 0.00, 4.21, 3.95, 3.85, 3.76, 
                 3.86, 3.66, 3.30, 3.35],
    "log2021": [7.60, 6.51, 6.57, 6.36, 6.25, 6.32, 0.00, 5.95, 5.85, 5.74, 
                 5.56, 5.50, 5.51, 5.45, 5.41, 5.22, 0.00, 5.29, 5.29, 5.24, 
                 4.95, 4.73, 4.76, 4.72, 4.42, 0.00, 4.24, 3.95, 3.88, 3.84, 
                 3.77, 3.67, 3.32, 3.34],
    "log2022": [7.54, 6.50, 6.41, 6.32, 6.20, 6.19, 0.00, 5.89, 5.75, 0.00, 
                 5.49, 5.49, 5.43, 0.00, 5.36, 5.24, 0.00, 5.23, 5.22, 5.16, 
                 4.89, 4.69, 4.67, 4.64, 4.42, 4.35, 4.21, 3.87, 3.83, 3.79, 
                 3.70, 3.63, 3.29, 3.23]
}
df = pd.DataFrame(data)

# Yılları tanımlayın
years = ["log2018", "log2019", "log2020", "log2021", "log2022"]

# "0" değerlerini NaN olarak değiştir
df[years] = df[years].replace(0, np.nan)

# Eksik verilerden arındırılmış bir DataFrame oluşturun
df_complete = df.dropna(subset=years)

# Verileri ölçeklendirme
scaler = StandardScaler()
X = scaler.fit_transform(df_complete[years])

# Elbow yöntemiyle ideal küme sayısını belirleme
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Elbow grafiğini çizme
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel("Küme Sayısı")
plt.ylabel("Inertia")
plt.title("Elbow Yöntemi ile Küme Sayısını Belirleme")
plt.show()

# Optimal küme sayısını belirledikten sonra K-Means uygulama
optimal_k = 4  # Grafikten belirlediğiniz optimal küme sayısını buraya girin
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
df_complete['Küme'] = kmeans.fit_predict(X)

# Ana veri setine küme etiketlerini geri ekleyin
df = df.merge(df_complete[['Ülkeler', 'Küme']], on='Ülkeler', how='left')

# Küme etiketleri
df['Küme Etiketi'] = df['Küme'].map({i: f"Küme {i+1}" for i in range(optimal_k)})

# Kümeleme sonuçlarını görselleştirme
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='log2018', y='log2022', hue='Küme Etiketi', data=df, palette='viridis', s=100
)
plt.title("Ülkelerin K-Means Kümeleme Analizi (2018-2022)")
plt.xlabel("2018 Log Varlık Değeri")
plt.ylabel("2022 Log Varlık Değeri")
plt.legend(title="Kümeler", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()

# Her bir kümenin ortalamasını hesaplama
print("Her kümenin ortalama değerleri:")
print(df.groupby('Küme')[years].mean())

# Her kümedeki ülkeleri listeleme
print("\nKüme Analizi Sonuçları:")
for i in range(optimal_k):
    cluster_countries = df[df['Küme'] == i]['Ülkeler'].values
    print(f"Küme {i + 1}: {', '.join(cluster_countries)}")

```

# Gelişmiş Kümeleme Analizi (PCA ile Boyut İndirgeme)¶
Verinin daha yüksek boyutlu olduğunu ve görselleştirme için boyut indirgeme ihtiyacı olduğunu düşündüğümüzde PCA (Principal Component Analysis) kullanılabilir. python

# Adım 1: PCA Bileşenlerini Doğru Şekilde Ekleyin
PCA bileşenlerini df_complete veri çerçevesine eklemek için doğru kodu kullanın. Eğer daha önceki işlemlerde bu adımı yapmadıysanız, aşağıdaki kodu çalıştırarak PCA bileşenlerini veri çerçevesine ekleyin:

```python
from sklearn.decomposition import PCA

# PCA ile boyut indirgeme işlemi
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X)  # X, ölçeklenmiş veri (StandardScaler uygulanmış)

# PCA bileşenlerini df_complete'e ekleyin
df_complete['PCA1'] = pca_components[:, 0]
df_complete['PCA2'] = pca_components[:, 1]
```

# Adım 2: Görselleştirme Kodunu Çalıştırın
Yukarıdaki adım tamamlandıktan sonra görselleştirme kodunu tekrar çalıştırabilirsiniz:

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='PCA1', y='PCA2', hue='Küme Etiketi', data=df_complete, palette='viridis', s=100
)

# Küme merkezlerini PCA boyutunda görselleştirme
centers_pca = pca.transform(kmeans.cluster_centers_)  # Küme merkezlerini PCA uzayına taşı
for i, center in enumerate(centers_pca):
    plt.text(center[0], center[1], f"Küme {i+1}", 
             fontweight='bold', fontsize=12, ha='center', va='center', color='red',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

plt.title("Ülkelerin Gelişmiş K-Means Kümeleme Analizi (PCA ile Boyut İndirgeme)")
plt.xlabel("PCA Bileşeni 1")
plt.ylabel("PCA Bileşeni 2")
plt.legend(title="Kümeler", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```
